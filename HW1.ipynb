```python
def createMatrix(rowCount, colCount):
    mat = []
    for i in range(rowCount):
        rowList = []
        for j in range(colCount):
            rowList.append(0)
        mat.append(rowList)

    return mat
```

```python
m1 = [[-3, -7], [-10, 6], [0, -6], [-1, -3]]
m2 = [[-8, 9], [4, -2]]
result = createMatrix(4, 2)
sum = 0
for row in range(len(m1)):
  for col in range(len(m1[row])):
    for row2 in range(len(m2[col])):
      result[row][col] = result[row][col] + m1[row][row2] * m2[row2][col]
print(result)
```
