H0: Среднесуточное количество переработанной нефти в разрезе на кварталы будет больше в 2021 относительно 2020 года

H1: Среднесуточное количество переработанной нефти в разрезе на кварталы не будет больше в 2021 относительно 2020 года

```python
#oilProd_2020 = [569.9, 468, 486.8, 500.6]
#oilProd_2021 = [512, 545, 555, 560]

oilRef_2020 = [244.7, 237.6, 218.4, 220.7]
oilRef_2021 = [227, 245, 203, 316]

alpha = 0.05

import scipy.stats as stats
statistics, pvalue = stats.ttest_rel(oilRef_2020, oilRef_2021)
print(statistics)
print(pvalue)
if (alpha > pvalue):
  print('Отвергаем гипотезу H0')
else:
  print('Не можем опровергнуть гипотезу H0')
```
