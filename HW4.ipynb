```python
A = np.array([[1, 2, 1], [-2, 1, 1]])
B = np.array([[3, 1], [-1, 0], [2, 4]])
C = np.array([[-2, 2], [1, -1], [1, 2]])

print(np.add(A, np.dot(2, B).reshape(2, 3)))
print(np.add(B, np.dot(2, C)))
```

```python
v1 = np.arange(1, 36 * 2, 2, 'int32')
print(v1)
```

```python
v2 = np.array([], 'int32')
for val in v1:
    if val % 3 == 0:
      v2 = np.append(v2, val)
print(v2)
```
